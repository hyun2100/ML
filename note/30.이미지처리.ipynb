### Image -> Numpy

```python
import glob
```

```python
numberofData = 300

img_size = 100

```

```python
# 배열 초기화로 만들기

fruits_data = np.zeros(numberofData * img_size * img_size).reshape(numberofData, img_size, img_size)
```

```python
#np.zeros([numberofData, img_size, img_size])
```

```python
i = 0

for file in sorted(glob.glob("../data/fruits/*.jpg")):
    img = np.array(Image.open(file), dtype= np.int32)
    fruits_data[i, :, :] = img
    i +=1
```

```python
fruits_data.shape
```

```python
# 이미지로 확인

import matplotlib.pyplot as plt


plt.imshow(fruits_data[0].reshape(100, 100), cmap='gray_r')

plt.show()
```
